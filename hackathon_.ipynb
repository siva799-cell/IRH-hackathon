```python
import pandas as pd
import matplotlib.pyplot as plt

final_data = pd.read_csv('final_food_delivery_dataset.csv')
print('final_data head:')
print(final_data.head())
```

```python
gold_members_df = final_data[final_data['membership'] == 'Gold']
print('\ngold_members_df head:')
print(gold_members_df.head())
```

```python
city_revenue = gold_members_df.groupby('city')['total_amount'].sum().reset_index()
city_revenue = city_revenue.sort_values(by='total_amount', ascending=False)
print('\ncity_revenue head:')
print(city_revenue.head())
```

```python
avg_order_value_by_cuisine = final_data.groupby('cuisine')['total_amount'].mean().reset_index()
print('\navg_order_value_by_cuisine head:')
print(avg_order_value_by_cuisine.head())
```

```python
user_total_orders = final_data.groupby('user_id')['total_amount'].sum().reset_index()
print('\nuser_total_orders head:')
print(user_total_orders.head())
```

```python
bins = [3.0, 3.6, 4.1, 4.6, 5.1]
labels = ['3.0-3.5', '3.6-4.0', '4.1-4.5', '4.6-5.0']
final_data['rating_range'] = pd.cut(final_data['rating'], bins=bins, labels=labels, right=False)
revenue_by_rating_range = final_data.groupby('rating_range', observed=False)['total_amount'].sum().reset_index()
revenue_by_rating_range = revenue_by_rating_range.sort_values(by='total_amount', ascending=False)
print('\nrevenue_by_rating_range:')
print(revenue_by_rating_range)
```

```python
revenue_combinations = []
gold_indian_revenue = final_data[(final_data['membership'] == 'Gold') & (final_data['cuisine'] == 'Indian')]['total_amount'].sum()
revenue_combinations.append({'Combination': 'Gold + Indian cuisine', 'Total Revenue': gold_indian_revenue})
gold_italian_revenue = final_data[(final_data['membership'] == 'Gold') & (final_data['cuisine'] == 'Italian')]['total_amount'].sum()
revenue_combinations.append({'Combination': 'Gold + Italian cuisine', 'Total Revenue': gold_italian_revenue})
regular_indian_revenue = final_data[(final_data['membership'] == 'Regular') & (final_data['cuisine'] == 'Indian')]['total_amount'].sum()
revenue_combinations.append({'Combination': 'Regular + Indian cuisine', 'Total Revenue': regular_indian_revenue})
regular_chinese_revenue = final_data[(final_data['membership'] == 'Regular') & (final_data['cuisine'] == 'Chinese')]['total_amount'].sum()
revenue_combinations.append({'Combination': 'Regular + Chinese cuisine', 'Total Revenue': regular_chinese_revenue})
revenue_df = pd.DataFrame(revenue_combinations)
print('\nrevenue_df head:')
print(revenue_df.head())
```

```python
final_data['order_date'] = pd.to_datetime(final_data['order_date'], format='%d-%m-%Y')
final_data['quarter'] = final_data['order_date'].dt.quarter
quarterly_revenue = final_data.groupby('quarter')['total_amount'].sum().reset_index()
quarterly_revenue = quarterly_revenue.sort_values(by='total_amount', ascending=False)
print('\nquarterly_revenue head:')
print(quarterly_revenue.head())
```

```python
avg_order_value_per_restaurant = final_data.groupby('restaurant_name_y')['total_amount'].mean().reset_index()
order_counts_per_restaurant = final_data.groupby('restaurant_name_y').size().reset_index(name='order_count')
restaurant_data = pd.merge(avg_order_value_per_restaurant, order_counts_per_restaurant, on='restaurant_name_y', how='inner')
restaurants_less_than_20_orders = restaurant_data[restaurant_data['order_count'] < 20]

restaurant_options = [
    'Grand Cafe Punjabi',
    'Grand Restaurant South Indian',
    'Ruchi Mess Multicuisine',
    'Ruchi Foods Chinese'
]
name_mapping = final_data[['restaurant_name_x', 'restaurant_name_y']].drop_duplicates()
options_mapped = name_mapping[name_mapping['restaurant_name_x'].isin(restaurant_options)]

highest_avg_order_from_options = pd.merge(
    options_mapped,
    restaurants_less_than_20_orders,
    on='restaurant_name_y',
    how='inner'
).sort_values(by='total_amount', ascending=False).iloc[0]

print('\nhighest_avg_order_from_options:')
print(highest_avg_order_from_options)
```

```python
num_distinct_users = final_data['user_id'].nunique()

hyderabad_revenue = final_data[final_data['city'] == 'Hyderabad']['total_amount'].sum()
rounded_hyderabad_revenue = round(hyderabad_revenue)

average_order_value_gold = gold_members_df['total_amount'].mean()
rounded_avg_order_value_gold = round(average_order_value_gold, 2)

orders_high_rating = final_data[final_data['rating'] >= 4.5]
num_orders_high_rating = len(orders_high_rating)

top_gold_revenue_city = city_revenue.iloc[0]['city']
orders_in_top_gold_city = gold_members_df[gold_members_df['city'] == top_gold_revenue_city]
num_orders_top_gold_city = len(orders_in_top_gold_city)

print('\nOther variables defined.')
```
